```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import itertools as it
```

```python
stock = pd.read_csv('data/DJIA_table.xz',index_col=0)
news = pd.read_csv('data/RedditNews.xz',index_col=0)
```

```python
stock.head()
```

```python
filter_out = news.groupby('Date').count().query('News != 25').index
```

```python
news.drop(labels=filter_out,axis=0, inplace=True)
```

```python
dates = news.index
prev = dates[0]
cnt = 0
cnts = []
for date in dates:
    if date == prev:
        cnt += 1
        cnts.append(cnt)
    else:
        cnt = 1
        cnts.append(cnt)
        prev = date
```

```python
news['Rank'] = cnts
```

```python
news.shape
```

```python
news.drop_duplicates(inplace=True)
```

```python
news.shape
```

```python
news = news.pivot(columns='Rank',values='News')
```
